```python
# importing libraries
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
from pytz import timezone
from datetime import datetime
import hvplot.pandas
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# load Heart data csv
Heart_2022 = pd.read_csv('Resources\heart_2022_no_nans.csv')
Heart_2022.head(-20)
```

```python
# Dropping BMI
Heart_2022.drop(columns=['BMI'], inplace=True)
Heart_2022.head(-20)
```

```python
columns_to_replace = ['PhysicalActivities', 'HadHeartAttack', 'HadAngina','HadStroke','HadAsthma','HadSkinCancer','HadCOPD','HadDepressiveDisorder','HadKidneyDisease','HadArthritis','HadDiabetes','DeafOrHardOfHearing','BlindOrVisionDifficulty','DifficultyConcentrating','DifficultyWalking','DifficultyDressingBathing','DifficultyErrands','ChestScan','AlcoholDrinkers','HIVTesting','FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear','CovidPos']
replacement_dict = {'Yes': 1, 'No': 0}
#Replace all the yes/no columns
for col in columns_to_replace:
    Heart_2022[col] = Heart_2022[col].replace(replacement_dict)
Heart_2022.head()
```

```python
replacement_dict = {'Male': 0, 'Female': 1}
Heart_2022['Sex'] = Heart_2022['Sex'].replace(replacement_dict)
#Replacing the sex column
Heart_2022.head()
```
